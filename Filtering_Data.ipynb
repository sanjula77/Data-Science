```python
import pandas as pd
```

```python
df_laptop = pd.read_csv('laptop_price.csv', encoding='ISO-8859-1')
df_laptop.head(3)
```

### Filter dataframe based on the condition

```python
# Find which raw have the "Apple" brand in the "Company" column
df_laptop['Company'] == 'Apple'
```

```python
# Filter dataframe based on the condition
df_laptop[df_laptop['Company'] == 'Apple']
```

```python
# Verify which companice are in the "Company" column with value_count()
df_laptop[df_laptop['Company'] == 'Apple'].value_counts('Company')
```

```python
# Find which raw don't have "HP" in the "Company" column
df_laptop['Company'] != 'HP'
```

```python
# Filter dataframe based on the condition 
df_laptop[df_laptop['Company'] != 'HP']
```

```python
# Value count without 'HP'
df_laptop[df_laptop['Company'] != 'HP'].value_counts('Company')
```

```python
# Find laptop with price over 2000 euros

df_laptop['Price_euros'] > 2000
```

```python
# Filter dataframe based on the condition

df_laptop[df_laptop['Price_euros'] > 2000]
```

#### Create a column based on 1 conditions:


```python
import numpy as np
```

```python
# Create an array based on price tiers

np.where(df_laptop['Price_euros'] > 2000, 'Expensive', 'Cheap')
```

```python
# Set it to a new column

df_laptop['price tiers'] = np.where(df_laptop['Price_euros'] > 2000, 'Expensive', 'Cheap')
```

```python
# Show first 5 rows

df_laptop.head()
```

```python
# Count value in price tiers column

df_laptop.value_counts('price tiers')
```
