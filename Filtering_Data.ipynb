#### Create a column based on 1 conditions:


```python
import numpy as np
```

```python
# Create an array based on price tiers

np.where(df_laptop['Price_euros'] > 2000, 'Expensive', 'Cheap')
```

```python
# Set it to a new column

df_laptop['price tiers'] = np.where(df_laptop['Price_euros'] > 2000, 'Expensive', 'Cheap')
```

```python
# Show first 5 rows

df_laptop.head()
```

```python
# Count value in price tiers column

df_laptop.value_counts('price tiers')
```

##### Filter dataframe based on multiple conditions

```python
# fine apple laptops that cost more than 2000 euros 
(df_laptop['Company'] == 'Apple') & (df_laptop['Price_euros'] > 2000)
```

```python
# Filter dataframe based on multiple conditions

df_laptop[(df_laptop['Company'] == 'Apple') & (df_laptop['Price_euros'] > 2000)]
```

#### Creating a conditional colunm for more than two

```python
# Create an array based on multiple price tiers
# Create 'conditional' and 'value' variables

conditional = [
    df_laptop['Price_euros'] > 3000,
    (df_laptop['Price_euros'] > 2000) & (df_laptop['Price_euros'] <= 3000),
    (df_laptop['Price_euros'] > 800) & (df_laptop['Price_euros'] <= 2000),
    df_laptop['Price_euros'] <= 800
]

value = ['Very Expensive', 'Expensive', 'Cheap', 'Very Cheap']
```

```python
# Set it to a new column

df_laptop['price tiers'] = np.select(conditional, value)
```

```python
# Show first 5 rows

df_laptop.head()
```

```python
# Count values inside 'price tiers' column

df_laptop.value_counts('price tiers')
```

#### isin() : Single filtering

```python
# select Apple or HP laptops

df_laptop['Company'].isin(['Apple', 'HP'])
```

```python
# Filter dataframe(first 5 rows) based on the condition

df_laptop[df_laptop['Company'].isin(['Apple', 'HP'])].head()
```

#### isin() : Multiple Filtering

```python
# Find Notebooks or ultrabooks From Apple or HP

filter_1 = df_laptop['Company'].isin(['Apple', 'HP']) 
filter_2 = df_laptop['TypeName'].isin(['Notebook', 'Ultrabook'])
```

```python
# Filter dataframe based on the condition

df_laptop[filter_1 & filter_2].head()
```

#### duplicated()

```python
# Find duplicates in 1 column/series

df_laptop.duplicated('laptop_ID')
```

```python
# Show elements that are duplicated in the 'laptop_ID' column

df_laptop[df_laptop.duplicated('laptop_ID')]
```

```python
# duplicates in two or more columns

duplicated = df_laptop.duplicated(['Product', 'TypeName', 'Inches'])
```

```python
# show all the values duplicated (except for the first - keep='first')

df_laptop[duplicated].sort_values(['Product', 'TypeName'])
```

#### Task: Find the cheapest and most expensive laptop of each company using the sort_values() and duplicated() methods (keep "first", "last" and False)

```python
# sort dataframe asending by company and price
# cheapest first and most expensive last
df_laptop = df_laptop.sort_values(['Company', 'Price_euros'])
df_laptop
```

```python
# check all value categories in "Company" column
df_laptop.value_counts('Company')
```

```python
# duplicated values in "Company" columns (default -> keep='first')
duplicated_first = df_laptop.duplicated('Company')
```

```python
# show dataframe with duplicated values in "Company" column
df_laptop[duplicated_first]

# keep='first'(cheapest laptop per company)
# show dataframe with non duplicated values in "Company" column
df_laptop[~duplicated_first]
df_laptop[~duplicated_first][['Company', 'Price_euros']]

# check all value categories
# df_laptops[~duplicated].value_counts('Company')
```

```python
# keep='last'
duplicated_last = df_laptop.duplicated('Company', keep='last')
# keep=False
duplicated_false = df_laptop.duplicated('Company', keep=False)
```

```python
# keep='last'(most expensive laptop per company)
# show dataframe with non duplicated values in "Company" column
df_laptop[~duplicated_last]
df_laptop[~duplicated_last][['Company', 'Price_euros']]
```

```python
# keep=False
# show dataframe with non duplicated values in "Company" column
df_laptop[~duplicated_false]
```

```python
# all elements in company column are repeated at least once
df_laptop.value_counts('Company')
```
